```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv('ex2.csv')
df.head()
```

```python
# Excercise 1 : Mapping Data
```

```python
size_dict = {'S':1,'M':2,'L':3,'XL':4}
```

```python
df['Size'] = df['Size'].map(size_dict)
```

```python
df
```

```python
Classlabel_dict = {'Class1':0, 'Class2':1}
df['Classlabel'] = df['Classlabel'].map(Classlabel_dict)
df
```

```python
# Excercise 2 : Inversing the Mapping
```

```python
size_inverse = {1:'S',2:'M',3:'L',4:'XL'}
```

```python
df['Size'] = df['Size'].map(size_inverse)
df
```

```python
# Excercise 3 : Label Encoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Size'] = le.fit_transform(df['Size'])
df
```

```python
df['Color'] = le.fit_transform(df['Color'])
```

```python
df
```
